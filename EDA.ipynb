```python
## 1) Which products are over-discounted but under-performing in revenue?

avg_discount=df['discount_percent'].mean()
print('Avg_Discount:',round(avg_discount,2))
```

```python
avg_revenue = df['total_revenue'].mean()
print('Avg_Revenue:',round(avg_revenue,2))
```

```python
result=df[(df['discount_percent'] > avg_discount) & (df['total_revenue'] < avg_revenue)][['product_id', 'product_category', 'discount_percent', 'total_revenue']]
```

```python
result
```

```python
## 2) How do product ratings relate to quantity sold?

rating_quantity=df.groupby('rating')['quantity_sold'].mean().reset_index()
rating_quantity
```

```python
## 3) Which categories generate high revenue with low sales volume?

df.groupby('product_category').agg(
          total_revenue=('total_revenue', 'sum'),
          total_quantity=('quantity_sold', 'sum')
      ).reset_index()
```
